```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
%matplotlib
```

```python
#           y-axis
plt.plot([1, 2, 3, 4])
# The default x-axis is used here. because we did not write.

# plt.plot([1, 1.2, 2, 2.2, 3, 3.2, 4])
# plt.plot([4, -4])
# plt.plot([-4, 4])
# plt.plot([4, 4])

plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.show()
```

```python
#           x-axis        y-axis
plt.plot([1, 2, 3, 4], [1, 2, 3, 9])
plt.show()
```

```python
plt.plot([1, 3, 4, 9], [-1, 8, 1, 2], 'ro',)
plt.show()
```

```python
array = np.arange(0, 5, 0.2)
#        x-axis y-axis HOW  | x-axis y-axis    HOW | x-axis  y-axis   HOW | x-axis  y-axis   HOW
plt.plot(array, array, 'r-.', array, array**2, 'bs', array, array**3, 'gp', array, array**4, 'y^')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
data = {
    # ----- array range -----
    'a': np.arange(500),
    # -----------------------
    
    # ----- random integer -----
    #                    START   length
    'b': np.random.randint(0, 50, 500),
    #                         END
    # --------------------------
    
    # ----- random normal -----
    # The numpy.random.randn() function creates an array of specified shape ...
    # ... and fills it with random values as per standard normal distribution.
    # https://www.geeksforgeeks.org/numpy-random-randn-python/
    # https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html
    
    # In the normal curve, the numbers are close to zero.
    'c': np.random.randn(500)
    # -------------------------
}

# data
```

```python
plt.scatter(data['a'], data['c'])
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
plt.show()
```
